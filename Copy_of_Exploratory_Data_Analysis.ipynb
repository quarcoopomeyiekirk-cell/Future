```python
# Cleaning the Data
```

```python
data.isnull().sum()
```

```python
anime = data[['score', 'scored_by', 'popularity', 'members', 'favorites', 'type']]
```

```python
anime.head()
```

```python
# Relationship analysis
```

```python
corelation = anime.corr
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

```python
sns.relplot(x='members', y='favorites', hue='type', data=anime)
```
